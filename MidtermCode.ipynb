```python
# 1번 이진검색 원문

from typing import Any, Sequence

def bin_search(a: Sequence, key: Any) -> int:
    """시퀀스 a에서 key와 일치하는 원소를 이진 검색"""
    pl = 0
    pr = len(a) - 1

    while True:
        pc = (pl + pr) // 2
        if a[pc] == key:
            return pc
        elif a[pc] < key:
            pl = pc + 1
        else:
            pr = pc - 1
        if pl > pr:
            break
    return -1

if __name__ == '__main__':
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num

    print('배열 데이터를 오름차순으로 입력하세요.')

    x[0] = int(input('x[0]: '))

    for i in range(1, num):
        while True:
            x[i] = int(input(f'x[{i}]: '))
            if x[i] >= x[i - 1]:
                 break

    ky = int(input('검색할 값을 입력하세요.: '))

    idx = bin_search(x, ky)
    if idx < 0:
        print('검색값을 갖는 원소가 존재하지 않습니다.')
    else:
        print(f'검색값은 x[{idx}]에 있습니다.')
```

```python
# 1번 문제 변형

def binary_search(array,target,start,end) -> int:  # 재귀를 위한 시작값과 끝값을 추가함
    if start > end:
        return -1   # 검색 실패시 값 반환

    mid = (start+end)//2

    if array[mid]==target:
        return mid  # 검색 성공, 재귀 종료

    elif array[mid] > target:
        return binary_search(array,target,start,mid-1)  # 범위 앞쪽 줄이기, 재귀로 앞부분 그대로 유지, 뒷부분은 중간값에서 -1을 뺀 값으로 설정

    else:
        return binary_search(array,target,mid+1,end)   # 범위 뒤쪽 줄이기, 재귀로 뒷부분 유지, 앞부분은 중간값 + 1로 실행
    

if __name__ == '__main__':
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num  # 원소 수가 num인 배열을 생성

    print('배열 데이터를 오름차순으로 입력하세요.')

    x[0] = int(input('x[0]: '))

    for i in range(1, num):
        while True:
            x[i] = int(input(f'x[{i}]: '))
            if x[i] >= x[i - 1]:
                 break

    ky = int(input('검색할 값을 입력하세요.: '))  # 검색할 ky를 입력

    idx = binary_search(x, ky, 0, len(x)-1)

    if idx < 0:
        print('검색값을 갖는 원소가 존재하지 않습니다.')
    else:
        print(f'검색값은 x[{idx}]에 있습니다.')
```

```python
## 2번 - 홀수번호 -> 검색 이용해 문제 출제하기

# 검색 : 원준이의 리스트에서 동덕이가 값을 맞추는 모험기
# 동덕이는 원준 리스트가 랜덤이기 때문에 선형검색을 사용한다
# 원준이의 리스트의 들어가는 원소 수 input 받는다
# 원준 리스트 값을 random(1,99)사이로 스스로 생성해낸다
# 검색 성공 : 원준에게서 승리함 프린트 + 리스트 위치 반환
# 검색 실패 : 원준에게 패배하였음 프린트

# 검색을 사용할 때
import random
def seq_search(a, key)-> int:
    """시퀀스 a에서 key값이 같은 요소를 선형 검색(for 문)"""
    for i in range(len(a)):
        if a[i] == key:
            return i  # 검색 성공(인덱스를 반환)
    return -1         # 검색 실패(-1을 반환)


if __name__ == '__main__':
    print('원준이의 리스트에 있는 동덕이가 값을 맞춘다')
    n = int(input("원준이의 리스트에 들어갈 원소의 개수를 입력하세요.: "))
    wj = [None] * n
    for i in range(0,n):
        wj[i] = random.randint(1,99)

    print(wj)

    dd = int(input("동덕이가 찾기를 도전하는 값을 입력하세요.: "))
    print(f"동덕이가 도전한 값 : {dd}")

    idx = seq_search(wj, dd)
    if idx < 0 :
        print("원준에게 패배하였습니다")
    else:
        print(f"값을 원준[{idx}]에서 찾았습니다! 원준에게 승리했습니다!")
```
