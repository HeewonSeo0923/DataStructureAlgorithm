```python
# 10진수 정수값을 입력받아 2-36 진수로 변환하여 출력하기

def card_conv(x: int, r: int) -> str:
    '''정수값 x를 r진수로 변환한 뒤 그 수를 나타내는 문자열로 반환'''

    d = '' # 변환 후
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWSYZ'

    while x > 0:
        d += dchar[x%r]
        x //= r

    return  d[::-1]   # 역순출력



if __name__ == '__main__':
    print('10진수를 n진수로 변환합니다.')

    while True:
        while True:
            no = int(input('변환할 값으로 음이 아닌 정수로 입력하세요.: '))
            if no > 0 :
                break

        while True:
            cd = int(input('어떤 진수로 변환할까요?: '))
            if 2 <= cd <= 36:
                break

        print(f'{cd}진수로는 {card_conv(no, cd)}입니다.')

        retry = input('한번 더 변환할까요?(Y ... 예 / N ... 아니오):')
        if retry in {'N', 'n'}:
            break
```

```python
# 1000 이하의 소수를 나열하기 (알고리즘 개선 1)

counter = 0                    # 나눗셈 횟수 카운트
ptr = 0                        # 이미 찾은 소수 개수
prime = [None] * 500           # 소수 지정 배열

prime[ptr] =2                  # 2소수이므로 초깃값
ptr += 1

for n in range(3, 1001, 2):    # 홀수만 지정
    for i in range(1, ptr) :   # 이미 찾은 소수로 나눔
        counter += 1
        if n % prime[i] == 0:
            break
    else:
        prime[ptr] = n
        ptr += 1

for i in range(ptr):
    print(prime[i])
print(f'나눗셈을 실행한 횟수: {counter}')
```
